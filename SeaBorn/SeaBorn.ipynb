```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mtcars=pd.read_csv('mtcars.csv')
```

```python
mtcars.head()
```

```python
mtcars.info()
```

```python
mtcars.describe()
```

```python
mtcars.shape
```

**Bar plot**

```python
sns.barplot(mtcars)
plt.show()
```

```python
sns.barplot(x='cyl',y='mpg',data=mtcars,color='r')
```

```python
sns.barplot(x=mtcars['cyl'],y=mtcars['mpg'],color='y')
plt.show()
```

```python

sns.barplot(x=mtcars['am'],y=mtcars['mpg'],color='b',palette='Set2' ) #Here palettes are color sets
plt.show()
```

```python
sns.barplot(x=mtcars['gear'],y=mtcars['hp'],color='b',palette='Set1' )
plt.show()
```

**Count plot x-axis**

```python
sns.countplot(x='cyl',data=mtcars,palette='Set1')
```

```python
sns.countplot(x='carb',data=mtcars,palette='Set2')
```

**Count plot y-axis**

```python
sns.countplot(y='gear',data=mtcars,palette='rocket')
```

**Groubing the count graph**

```python
sns.countplot(x='gear',hue='cyl',data=mtcars,palette='Set1')
```

**DistPlot**

```python
sns.distplot(mtcars.mpg, bins=10,color='g')
```

**Heatmap**

```python
mtcars.drop(axis=1,columns='model',inplace=True) #actually i had dropped this column
```

```python
sns.heatmap(mtcars.corr() , cbar=True, linewidths=0.5)
```

**Scatter Plot**

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris)
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species')
```

**Pair Plots**

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue="species")
```

**Lm plot**

```python
sns.lmplot(x='petal_length',y='petal_width',data=iris)
```

```python
sns.lmplot(x='petal_length',y='petal_width',data=iris,hue='species',markers=['o','*','^'])
```

**Box Plot**

```python
sns.boxplot(x='species',y='sepal_width',data=iris,hue='species')
```

**Violin Plot**

```python
sns.violinplot(x='species',y='sepal_width',data=iris,hue='species')
```

**Line Plot**

```python
sns.lineplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
```
